```python
def find_characters(input_string):
    final_string = ""
    
   
    pairs = input_string.split(',')
    
    for pair in pairs:
        name, code = pair.split(':')
        name_length = len(name)
        max_digit = -1
        
       
        for digit in code:
            digit_value = int(digit)
            if digit_value <= name_length:
                max_digit = max(max_digit, digit_value)
        
        
        if max_digit != -1:
            final_string += name[max_digit - 1] 
        else:
            final_string += "X"
    
    return final_string

input_string = "Abhishek:12348,Anirban:23745,Amit:3456,Manna:999"
output = find_characters(input_string)
print(output)  
```
