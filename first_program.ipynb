```python
adjectives=['wet','proud','sleepy','smelly','fatty','red','fluffy','brave']
nouns=['apples','dinosour','ball','goat','dragon','panda']
print("Welcome to the password picker game:")
while True:
    adjective=random.choice(adjectives)
    noun=random.choice(nouns)
    number=random.randrange(1,100)
    scr=random.choice(string.punctuation)
    password=adjective+noun+str(number)+scr
    print("Your new password is: ",password)
    response=input(""""Would you like another password...?
                    Type y for Yes and n for No    :""")
    if response=='n':
        break
```

### Nine lives

```python
import random
```

```python
words=['shirt','teach','pizza','fairy','plane']
secret_word=random.choice(words)
print(secret_word)
clue=list('?????') #clue variable
```

```python
heart_symbol=u'\u2764'
print(heart_symbol)
```

```python
def update_clue(secret_word,guessed_letter,clue):
    index=0
    while index<len(secret_word):
        if guessed_letter==secret_word[index]:
            clue[index]=guessed_letter
        index=index+1
```

```python
guess_word_correctly=False
lives=9
while lives>0:
    print(clue)
    print("Lives left "+ heart_symbol*lives)
    guess=input("Guess a letter or the whole word..")
    
    if guess==secret_word:
        guess_word_correctly=True
        break
    if guess in secret_word:
        update_clue(secret_word, guess, clue)
    else:
        print("Incorrect you loss the life  .... ")
        lives=lives-1
        
if guess_word_correctly:
    print("You won the game....!")
else:
    print("You loss the game....!")
```
